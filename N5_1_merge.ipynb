## Combinaçao de tabelas (Merge)

Principalmente em estruturas de bancos de dados relacionais, como o *SQL Server* da Microsoft e semelhantes, é muito comum fazermos a combinação de tabelas de diferentes formas. O *merge* é uma combinação de tabelas através de uma variável *chave* que relaciona as duas tabelas.

Por exemplo, imagine que temos duas tabelas:

- ```df1``` que contém informações do cliente, em que cada linha representa um cliente, e cada coluna uma informação, como um identificador desse cliente (```ID```), sexo, idade e salário.
- ```df2``` contém informações de compra, em que cada linha representa uma compra, e as colunas as informações dessa compra, como o cliente que fez a compra, o valor comprado no mês 1, e o valor comprado no mês 2.

Uma tarefa bastante natural é querer unir estas informações em uma única tabela.

````python
import pandas as pd
````

````python
df1 = pd.DataFrame({
    'ID': ['Cliente_1', 'Cliente_2', 'Cliente_3'],
    'Sexo': ['M', 'F', 'F'],
    'Idade': [27, 30, 24],
    'Salario': [5000, 7000, 2700]
})
df1
````

````python
df2 = pd.DataFrame({
    'ID': ['Cliente_2', 'Cliente_3', 'Cliente_4'],
    'Compras_M1': [200, 300, 1000],
    'Compras_M2': [800, 300, 2000]
})
df2
````

### Left join <a name="left_2"></a>
[Voltar ao índice](#Contents)

<div>

</div>

Queremos a tabela da "esquerda" como está, e agregar a ela uma ou mais variáveis da tabela da "direita", identificando os registros por uma *chave*.

- As linhas da tabela da "esquerda" sem correspondência, ficam missing nas novas variáveis
- As linhas da "direita" sem correspondência na esquerda são ignoradas.

````python
df1.merge(df2, on='ID', how='left')
````

### Right<a name="right_2"></a>    
[Voltar ao índice](#Contents)

<div>

</div>
Semelhante ao *left join*, mas ficamos cm as variáveis da tabela da "direita". Você sempre pode fazer o mesmo com um *left join* invertendo as bases da "esquerda" e da "direita".

"O *right join* é um *left join* de ponta cabeça."

````python
df1.merge(df2, on='ID', how='right')
# df2.merge(df1, on='ID', how='left')
````

````python
df2.merge(df1, on='ID', how='left')
````

### Inner join<a name="inner"></a>
     
[Voltar ao índice](#Contents)

<div>

</div>

O *inner join* mantém somente as linhas cuja chave aparece em ambas as tabelas. Como resultado, não produz valores não preenchidos por não constar em uma das duas tabelas.

````python
df1.merge(df2, on='ID', how='inner')
````

### Full outer join <a name="full"></a>
[Voltar ao índice](#Contents)

<div>

</div>

Depois que entendemos a lógica fica fácil: o *full join* retornará todos os registros, pareando as variáveis pela chave, que estejam na tabela da "direita" **e** na tabela da "esquerda"

- As variáveis da tabela da "esquerda" ficarão missing para os registros sem correspondência na tabela da "direita".
- As variáveis da tabela da "direita" ficarão missing para os registros sem correspondência na tabela da "esquerda".

No exemplo abaixo, o cliente 4 não tem correspondente na tabela 2, e o cliente 3 não tem correspondente na tabela 1. Veja os NaNs produzidos.

````python
df1.merge(df2, on='ID', how='outer')
````

````python
df1.merge()
````

### Merge com linhas repetidas

Devemos ter muita atenção ao fazer um *merge*, em especial quando temos linhas repetidas - devemos entender o que são as repetições, e se são legítimas. Muitas vezes sim, o caso contrário pode ser necessário tratamento prévio.

**Cuidado**: chaves repetidas dos dois lados provavelmente significam erro.

````python
df_esq = pd.DataFrame({'cliente': [1, 2, 3, 4],
                       'score': [980, 890, 950, 540]})
df_esq
````

````python
df_dir = pd.DataFrame({'cliente': [1, 1, 2, 3],
                       'cartão': ['a', 'b', 'a', 'b'],
                       'valor': [100, 200, 150, 250]})
df_dir
````

````python
df_esq.merge(df_dir, on='cliente', how='left')
````

## Concatenando tabelas <a name="5"></a>
[Voltar ao índice](#topo)

Podemos querer concatenar tabelas ao longo dos dois eixos definidos no Pandas - ```index``` e ```columns```. Concatenar ao longo do ```index``` é o que chamamos de *append* (ou fazer apêndice em português).

Adicionar dados "lateralmente" é um caso particular do *merge*.

<div></div>

A operação de empilhar tabelas é usualmente chamada de *append*. Ela pode ser realizada com o comando ```pd.concat()``` ou com o método ```append()```. Ambos acabam produzindo resultados equivalentes, alterando essencialmente a forma da instrução.

### Função ```concat```

A função ```pd.concat()``` concatena bases de dados de forma simples ao longo do eixo especificado. O eixo padrão é ```axis = 0```, produzindo uma tabela com a quantidade de linhas igual a soma das linhas das tabelas concatenadas. Usar ```axis = 1``` vai resultar numa tabela com o número de colunas igual à soma das colunas.

````python
dfa = pd.DataFrame({
    'ID': ['Cliente_1', 'Cliente_2'],
    'Sexo': ['M', 'F'],
    'Idade': [27, 30]
})
dfa
````

````python
dfb = pd.DataFrame({
    'ID': ['Cliente_3', 'Cliente_4'],
    'Sex': ['F', 'M'],
    'Idade': [24, 35]
})
dfb
````

````python
dfc = pd.DataFrame({
    'ID': ['Cliente_5', 'Cliente_6', 'Cliente_1'],
    'Sexo': ['M', 'F', 'F'],
    'Idade': [40, 50, 85]
})
dfc
````

O concat une tabelas horizontal ou verticalmente, dependendo do parâmetro ```axis```. Configurando ```axis=0``` temos uma concatenação vertical conforme a figura:


````python
pd.concat([dfa, dfb.rename(columns={'Sex':'Sexo'}), dfc]
          # , axis = 0
          , ignore_index = True
         )
````

A opção ```axis=1``` concatena horizontalmente. Um efeito semelhante a um ```.merge``` com chaves já pareadas e em mesma ordem. Em geral é preferível utilizar o ```.merge``` explicitando a chave.


Repare que no exemplo abaixo ficamos com variáveis com nomes repetidos, o que não costuma ser uma boa ideia.

````python
pd.concat([dfa, dfb, dfc]
          , axis = 1
         )
````

### Método ```.append()``` <a name="append"></a>

O método ```append``` é um "atalho" para a função concat.

```df1.append([df2])``` é equivalente a ```pd.concat([df1,df2], axis = 0)```.

Para empilhar os dataframes df1, df2 e df3:

````python
dfa.append([dfb, dfc])
````

É equivalente a:

````python
base_empilhada = pd.concat([dfa,dfb, dfc])
base_empilhada
````
